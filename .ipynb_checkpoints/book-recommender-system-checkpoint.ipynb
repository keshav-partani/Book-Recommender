## Popularity Based Recommendation System

```python
ratings_with_name = ratings.merge(books, on="ISBN")
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating': 'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].agg('mean').reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popularity_df = avg_rating_df.merge(num_rating_df, on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False)
popularity_df = popularity_df.head(50)
popularity_df
```

```python
popularity_df = popularity_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df
```

```python
popularity_df = popularity_df[['Book-Title', 'Book-Author','Year-Of-Publication','Image-URL-M', 'num_ratings', 'avg_ratings']]
popularity_df
```

# Collaborative Filtering Based Recommender

```python
x = ratings_with_name.groupby("User-ID").count()['Book-Rating'] > 200
book_reader_user = x[x].index
```

```python
filter_rating = ratings_with_name[ratings_with_name['User-ID'].isin(book_reader_user)]
```

```python
y = ratings_with_name.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_books)]
final_rating
```

```python
pt = final_rating.pivot_table(index="Book-Title", columns='User-ID', values="Book-Rating")
pt
```

```python
pt.fillna(0, inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    distance = similarity_score[index]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x: x[1], reverse=True)[1:20]
    
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```
